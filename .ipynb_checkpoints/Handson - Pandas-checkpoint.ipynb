# A smart skill to master in Pandas - Pivot/Unpivot 

```python
notgtdataonly.columns
```

```python
notgtdataonly.to_csv('NOTGT.csv')
```

```python
notgtdataonly=notgtdataonly.head(20000)
```

```python
import numpy as np

pivot_table2=pd.pivot_table(notgtdataonly, values='Sales', aggfunc=np.sum,
                            index=['Customer'],
                            columns=['Division'],
                            fill_value=100000)
```
